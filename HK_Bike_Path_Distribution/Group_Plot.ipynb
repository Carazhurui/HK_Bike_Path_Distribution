### 7 Word cloud of policy

Since there were so many news and policies regarding bike that it was impossible for us to read and analysis the file by file and line by line, to get a general idea of policies about bicycle, we chose to take web scraping approach to scrape the most recent 100 articles published in GovHK and draw a word cloud based on their content.  
'For' loop function was used to help us automatically collect data from page one to page ten.We collected the title and the url of each article and put all of them in a dictionary which was further saved as a csv file.  
Then we use 'for' loop again to get all the 'text' from each url. After removing all the extra charactors, our document contained Chinese only.  
Before drawing the word cloud, we used 'jieba' function to get our document prepared. Finally, we plotted a word cloud showing the most frequent used 200 words in the most recent 100 bike related polices.

#### 7.1 Web scraping

```python
import requests 
import bs4 as bs 
import pandas as pd
import numpy as np
```

```python
title = []
link = []

for i in range(10):
    source = requests.get('https://www.search.gov.hk/result?ui_lang=zh-hk&proxystylesheet=ogcio_home_adv_frontend&output=xml_no_dtd&ui_charset=utf-8&a_submit=false&query=%E5%8D%95%E8%BD%A6%20%E7%AB%8B%E6%B3%95%E4%BC%9A&ie=UTF-8&oe=UTF-8&site=gia_home&tpl_id=stdsearch&gp=0&gp0=gia_home&gp1=gia_home&p_size=10&num=10&doc_type=all&as_filetype=&as_q=&as_epq=&is_epq=&as_oq=&is_oq=&as_eq=&is_eq=&r_lang=&lr=&web=this&sw=1&txtonly=0&rwd=0&date_v=%23-1&date_last=%23-1&s_date_year=2019&s_date_month=01&s_date_day=01&e_date_year=2019&e_date_month=12&e_date_day=04&last_mod=&sort=date%3AD%3AL%3Ad1&page='+str(i+1))
    source.encoding = 'utf-8'
    soup = bs.BeautifulSoup(source.content , features = 'html.parser')
    news_container = soup.find('div',{'class':'searchListing'})
    
    try:
        for news in news_container.find_all('div',{"class":"title"}):
            link_content = news.text.replace('\n','')
            title.append(link_content)
    except:
        news.append('None')

    try:
        for news in news_container.find_all('div',{"class":"link"}):
            link_content = news.find('a').get('href')
            link.append(link_content)
    except:
        news.append('None')
    
print(title)
```

```python
dic = {
    'Title':title,
    'Link':link
}
```

```python
HK_news = pd.DataFrame.from_dict(dic,orient='index').transpose()
print(HK_news.info())
HK_news
```

```python
HK_news.index = np.arange(1, len(title) + 1 ) # index from 1
HK_news.head(5)
```

```python
HK_news.to_csv('HK_bike_news.csv',encoding='utf-8')
```

```python
content = []

import re
 
for i in link:
    
    content_source = requests.get(i)
    content_source.encoding = 'utf-8'
    
    soup = bs.BeautifulSoup(content_source.content , features = 'html.parser')
    news_content = soup.find('body').text
    
    chinese = '[\u4e00-\u9fa5]+'
    news_content_chi = re.findall(chinese,news_content)
    
    for j in range(len(news_content_chi)):
        content.append(news_content_chi[j])
    
print(len(content))
print(content)
```

```python
import jieba

content_str = "".join(content)
words_list = jieba.lcut(content_str)
words_str = "".join(words_list)

print(type(words_str))
print(words_str)
```

```python
with open('news content.txt','w') as news_file:
    news_file.write(words_str)
```

```python
from wordcloud import WordCloud
from random import randint
import imageio
from scipy.misc import imread
import jieba
```

```python
def random_color_func(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
        h  = randint(120,250)
        s = int(100.0 * 255.0 / 255.0)
        l = int(100.0 * float(randint(60, 120)) / 255.0)
        return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
word_cloud_source = jieba.lcut(words_str)

wc = WordCloud(
    background_color='white',
    width = 2000,
    height = 1000,
    max_words= 200,
    max_font_size= 300,
    min_font_size=5,
    random_state=20,
    mask = M,
    color_func=random_color_func,
    font_path = 'Arial Unicode.ttf')

word_cloud = wc.generate(','.join(word_cloud_source))
word_cloud.to_file("news content.jpg") 
```
